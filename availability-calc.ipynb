```python
%matplotlib inline

import numpy
import matplotlib.pyplot as plt
```

# Availability Calculator

## Parameters

```python
MEHTOD = 'quorum' # Values = [ quorum ]
N = 3
```

```python
class QuorumKernel(object):
    @property
    def title(self):
        return "N=%d Quorum" % self._N
    
    def __init__(self, N=3):
        self._N = N
    
    def compute(self, a_node):
        return 3 * pow(a_node, 2) - 2 * pow(a_node, 3)

class Plotter(object):
    label_fontsize = 16
    title_fontsize = 20
    legend_fontsize = 20
    
    def __init__(self, kernel):
        self._kernel = kernel
    
    def plot(self, a_node, title=None):
        a_cluster = numpy.array(list(map(self._kernel.compute, a_node)))
        plt.figure(figsize=(6,4), )
        plt.plot(
            a_node, 
            a_cluster,
            label=self._kernel.title)

        plt.plot(
            a_node, 
            a_node,
            label="Reference")

        plt.legend(loc=0, fontsize=self.legend_fontsize)
        plt.xlabel("A-Node", fontsize=self.label_fontsize)
        plt.ylabel("A-Cluster", fontsize=self.label_fontsize)
        plt.axis('tight')

        plt.title(
            title or 'Availability (%.2f - %.2f)' % (min(a_node), max(a_node)),
            fontsize=self.title_fontsize
        )
```

```python
plotter_quorum_3 = Plotter(QuorumKernel(N=3))
plotter_quorum_3.plot(numpy.linspace(0.5, 1.0, 50))
```
